## FLOPS - Fast equivalent layer Vs BCCB

```python
N = np.arange(10000.,1000000,10000)
it = 50
old = it*(3*N + 2*N**2)
```

```python
k = 5.
new = it*(27*N + k*8*N*np.log2(4*N))
```

```python
plt.figure(figsize=(10,6))

plt.semilogy(N, old, 'ko:', alpha=1, markersize=8, label='Siqueira et al')
plt.semilogy(N, new, 'b^--', alpha=1, markersize=8, label='Our method')
#plt.axis('scaled')
#plt.xlim(np.min(N)-1,np.max(N)+1)
plt.ylim(10**7,10**15)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Number of observations', fontsize=18)
plt.ylabel('Floating points', fontsize=18)
plt.grid(True)
plt.tight_layout(True)
plt.legend(loc='upper left', fontsize=14)
#plt.savefig('figures/float.png', dpi= 500)
```
